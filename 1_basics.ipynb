### Задание 1. Загрузка данных
Изучить представленный набор данных на основе описания его столбцов, загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

```python
#загружаю 8 столбцов из файла
df = pd.read_csv('1_basics/horse_data.csv', usecols=[0,1,3,4,5,6,10,22], 
                 names = ['surgery', 'age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome'],
                 header=0, na_values = '?')
df.head()
```

```python
#смотрю инфо
df.info()
df.head()
```

```python
#изменяю тип данных для столбцов
df['surgery'] = df['surgery'].astype('Int64')
df['temperature of extremities'] = df['temperature of extremities'].astype('Int64')
df['pain'] = df['pain'].astype('Int64')
df['outcome'] = df['outcome'].astype('Int64')
```

### Задание 2. Первичное изучение данных
Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
#299 entries
df.describe()
```

```python
# Столбцы с фикс. целыми значениями
# age: должен принимать значения 1 или 2: но 2 нет, предполагаю, что все 9 - на самом деле 2
df[df['age'] > 2]
df['age'].unique()
df['age'].replace(9, 2, inplace = True)
```

```python
# в других столбцах все значения как в описании, больше выбросы не ищу
df['surgery'].unique()
df['temperature of extremities'].unique()
df['pain'].unique()
df['outcome'].unique()
```

```python
# выбросы в столбцах с float
# температура - выбросы по формуле есть, но не кажутся совсем дикими
q1 = df['rectal temperature'].quantile(0.25)
q3 = df['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
df[(df['rectal temperature'] < lower_bound) | (df['rectal temperature'] > upper_bound)].sort_values('rectal temperature')
```

```python
#пульс
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
print(lower_bound)
print(upper_bound)
df[(df['pulse'] < lower_bound) | (df['pulse'] > upper_bound)].sort_values('pulse')

# убираю эти выбросы, тк не вижу зависимости от других криетриев = их мало. Перезаполн в задании 3
df['pulse'] = np.where(df['pulse'] > upper_bound, np.nan, df['pulse'])
```

```python
# respiratory rate
# normal rate is 8 to 10
# usefulness is doubtful due to the great fluctuations
q1 = df['respiratory rate'].quantile(0.25)
q3 = df['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
print(lower_bound)
print(upper_bound)
df[(df['respiratory rate'] < lower_bound) | (df['respiratory rate'] > upper_bound)].sort_values('respiratory rate')

# не трогаю выбросы, тк судя по описанию все данные в столбце под вопросом
```

### Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
df.info()
```

```python
# для столбца outcome дозполнять 1 строку не буду, удалю эту запись (в ней же был неизвестен признак surgery)
df.dropna(subset=['outcome'], inplace = True)
```

```python
# дозаполнять данные с большим кол-вом пропусков также не хочется (все 5)
df.dropna(thresh=5, inplace = True)
```

```python
# ок, удалилось 11 строк
# надо дозаполнить данные в 5 столбцах
df.info()
```

```python
# respiratory rate
# из описания данных: usefulness is doubtful due to the great fluctuations
# заполняю медианой без привязки к другим признакам
df['respiratory rate'].fillna(df['respiratory rate'].median(), inplace=True)
```

```python
# temperature of extremities и pain
# заполняю модой 
df['temperature of extremities'].fillna(df['temperature of extremities'].mode()[0], inplace=True)
df['pain'].fillna(df['pain'].mode()[0], inplace=True)
```

```python
# pulse
# заполняю медианой в зависимости от pain
df[df['pulse'].isnull()].sort_values('pulse')
print(df.groupby('pain')['pulse'].median())
df['pain'].value_counts()

df['pulse'].fillna(df.groupby('pain')['pulse'].transform('median'), inplace=True)
```

```python
# rectal temperature
# сложно привязать к другим признакам, тк может меняться в разныне стороны в зависимости от того, как протекает болезнь
# взаимосвязь есть со столбцом temperature of extremities: hot extremities should correlate with an elevated rectal temp
# заполняю медианой в зависимости от temperature of extremities
df['rectal temperature'].fillna(df.groupby('temperature of extremities')['rectal temperature'].transform('median'), inplace=True)
```
